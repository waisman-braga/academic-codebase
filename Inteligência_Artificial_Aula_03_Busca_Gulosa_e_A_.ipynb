```python
import random
```

```python
M = [
      [1, 2, 3],
      [8, 0, 4],
      [7, 6, 5],
    ]

def imprimir(M):
  for v in M:
    print(v)
  print('---')

imprimir(M)
```

```python
def get_zero(M):
  for i in range(0,len(M)):
    for j in range(0,len(M)):
      if M[i][j] == 0: return i, j
  return None

def copiar(M):
  Z = [row[:] for row in M]
  return Z


def cima(M):
  i,j = get_zero(M) # posicao zero / espaco branco
  novo_i = i-1
  if novo_i < 0: return None
  novo_j = j

  Z = copiar(M)

  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0
  return Z

def baixo(M):
  i,j = get_zero(M) # posicao zero / espaco branco
  novo_i = i+1
  if novo_i > len(M)-1: return None
  novo_j = j

  Z = copiar(M)

  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0
  return Z

def esquerda(M):
  i,j = get_zero(M) # posicao zero / espaco branco
  novo_j = j-1
  if novo_j < 0: return None
  novo_i = i

  Z = copiar(M)

  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0
  return Z

def direita(M):
  i,j = get_zero(M) # posicao zero / espaco branco
  novo_j = j+1
  if novo_j > len(M)-1: return None
  novo_i = i

  Z = copiar(M)

  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0
  return Z

Z = direita(M)
imprimir(M)
print('')
imprimir(Z)
```

```python
class Node:
    def __init__(self, estado, pai, movimento, profundidade):
        self.estado = estado
        self.pai = pai
        self.movimento = movimento
        self.profundidade = profundidade
        self.custo = None # armazenamento do custo


def expandir_node(node_pai):

    nodes_filhos = []

    node_filho_C = Node(cima(node_pai.estado), node_pai, 'C', node_pai.profundidade+1 )
    if node_filho_C.estado != None: nodes_filhos.append(node_filho_C)

    node_filho_B = Node(baixo(node_pai.estado), node_pai, 'B', node_pai.profundidade+1)
    if node_filho_B.estado != None: nodes_filhos.append(node_filho_B)

    node_filho_E = Node(esquerda(node_pai.estado), node_pai, 'E', node_pai.profundidade+1)
    if node_filho_E.estado != None: nodes_filhos.append(node_filho_E)

    node_filho_D = Node(direita(node_pai.estado), node_pai, 'D', node_pai.profundidade+1)
    if node_filho_D.estado != None: nodes_filhos.append(node_filho_D)

    return nodes_filhos


node_inicial = Node(M, None, None, 0)
nodes_filhos = expandir_node(node_inicial)

imprimir(node_inicial.estado)
for n in nodes_filhos:
  imprimir(n.estado)
```

```python
estado_final = [
      [1, 2, 3],
      [8, 0, 4],
      [7, 6, 5],
    ]

def embaralhar(estado_final):

  Z = copiar(estado_final)

  for i in range(0,30):
    operacao = random.randint(1, 4)

    if operacao==1:
      C = cima(Z)
      if C != None:
        #print('C')
        Z = C

    if operacao==2:
      B = baixo(Z)
      if B != None:
        #print('B')
        Z = B

    if operacao==3:
      E = esquerda(Z)
      if E != None:
        #print('E')
        Z = E

    if operacao==4:
      D = direita(Z)
      if D != None:
        #print('D')
        Z = D

  return Z

estado_inicial = embaralhar(estado_final)

print('Estado Inicial:')
imprimir(estado_inicial)
print('Estado Final:')
imprimir(estado_final )


```

```python
import numpy as np
def heuristica(estado_atual, estado_final): #heuristica para quebra-cabeca-8 - num pecas fora do lugar
  counter=0
  for i in range(0,len(estado_atual)):
      for j in range(0,len(estado_atual)):
        if estado_atual[i][j] != estado_final[i][j]:
          counter += 1 # fora do lugar

  return counter

def get_pos(M,bloco):
  for i in range(0,len(M)):
    for j in range(0,len(M)):
      if M[i][j] == bloco: return i, j
  return None

def heuristica2(estado_atual, estado_final): #heuristica para quebra-cabeca-8 - dist manhatam
  dist = 0
  for bloco in range(1,9):
    x_final, y_final = get_pos(estado_final,bloco)
    x_atual, y_atual = get_pos(estado_atual,bloco)
    dist += abs(x_atual - x_final) + abs(y_atual - y_final)

  return dist
```

```python
def busca_gulosa(estado_inicial, estado_final, profundidade_maxima=10):

    node_inicial=Node(estado_inicial,None,None,0)
    fila_p=[]
    fila_p.append(node_inicial)
    node_atual=fila_p.pop(0)

    while(node_atual.estado!=estado_final):

        node_filhos=expandir_node(node_atual)

        for node_filho in node_filhos:
          if node_filho.profundidade <= profundidade_maxima:
            h = heuristica(node_filho.estado, estado_final)
            node_filho.custo = h
            fila_p.append(node_filho)


        fila_p.sort(key =lambda node: node.custo) # ordena de forma crescente pela heuristica
        node_atual=fila_p.pop(0)
        print('Visitando')
        imprimir(node_atual.estado)
        print(node_atual.custo)


    caminho_solucao=[]
    while(node_atual.pai!=None):
        caminho_solucao.insert(0,node_atual)
        node_atual=node_atual.pai
    return caminho_solucao


```

```python
# Estado Inicial:
# [3, 8, 4]
# [1, 6, 2]
# [7, 0, 5]

imprimir(estado_inicial)
solucao = busca_gulosa(estado_inicial, estado_final)


for s in solucao:
  print(s.movimento)
  imprimir(s.estado)
```

```python
def busca_a_estrela(estado_inicial, estado_final):

    node_inicial=Node(estado_inicial,None,None,0)
    fila_p=[]
    fila_p.append(node_inicial)
    node_atual=fila_p.pop(0)

    nos_visitados = 0

    while(node_atual.estado!=estado_final):

        node_filhos=expandir_node(node_atual)
        nos_visitados+=1
        for node_filho in node_filhos:

            g = node_filho.profundidade
            h = heuristica2(node_filho.estado, estado_final)

            node_filho.custo = g+h
            fila_p.append(node_filho)


        fila_p.sort(key =lambda node: node.custo) # ordena de forma crescente pelo custo
        node_atual=fila_p.pop(0)
        #print('Visitando')
        #imprimir(node_atual.estado)
        #print(node_atual.custo)

    print('Visitados: ',nos_visitados)
    caminho_solucao=[]
    while(node_atual.pai!=None):
        caminho_solucao.insert(0,node_atual)
        node_atual=node_atual.pai
    return caminho_solucao


```
