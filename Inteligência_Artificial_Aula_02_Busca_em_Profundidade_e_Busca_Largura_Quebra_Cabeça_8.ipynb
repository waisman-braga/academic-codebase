Exemplo de estado para o Quebra-Cabeça 8

```python
M = [
    [1, 2, 3],
    [8, 0, 4],
    [7, 6, 5],
]

def imprimir(M):
  for v in M:
    print(v)
  print('----')

imprimir(M)
```

Identificar o espaço em branco (valor zero) de um estado e fazendo operações de movimentação Cima, Baixo, Esquerda e Direita

```python
def get_zero(M):
  for i in range(0,len(M)):
    for j in range(0,len(M)):
      if M[i][j] == 0: return i, j
  return None

def copiar(M):
  Z = [row[:] for row in M]
  return Z


def cima(M):
  i,j = get_zero(M)
  novo_i = i-1
  if novo_i < 0: return None
  novo_j = j

  Z = copiar(M)
  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0

  return Z

def baixo(M):
  i,j = get_zero(M)
  novo_i = i+1
  if novo_i > len(M)-1: return None
  novo_j = j

  Z = copiar(M)
  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0

  return Z

def esquerda(M):
  i,j = get_zero(M)
  novo_j = j - 1
  if novo_j < 0: return None
  novo_i = i

  Z = copiar(M)

  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0

  return Z

def direita(M):
  i,j = get_zero(M)
  novo_j = j + 1
  if novo_j > len(M)-1: return None
  novo_i = i

  Z = copiar(M)

  Z[i][j] = Z[novo_i][novo_j]
  Z[novo_i][novo_j] = 0

  return Z

imprimir(M)
Z = baixo(M)
imprimir(Z)
```

Representando cada estado por meio de um Node.

Vamos usar cada Node como um nó na árvore de busca.

```python
class Node:
  def __init__(self, estado, pai, movimento, profundidade):
    self.estado = estado
    self.pai = pai
    self.movimento = movimento
    self.profundidade = profundidade

def expandir_node(node_pai):

  nodes_filhos = []

  node_filho_C = Node( cima(node_pai.estado), node_pai, 'CIMA', node_pai.profundidade+1 )
  if node_filho_C.estado != None: nodes_filhos.append(node_filho_C)

  node_filho_B = Node( baixo(node_pai.estado), node_pai, 'BAIXO', node_pai.profundidade+1 )
  if node_filho_B.estado != None: nodes_filhos.append(node_filho_B)

  node_filho_E = Node( esquerda(node_pai.estado), node_pai, 'ESQUERDA', node_pai.profundidade+1 )
  if node_filho_E.estado != None: nodes_filhos.append(node_filho_E)

  node_filho_D = Node( direita(node_pai.estado), node_pai, 'DIREITA', node_pai.profundidade+1 )
  if node_filho_D.estado != None: nodes_filhos.append(node_filho_D)

  return nodes_filhos

estado = [
    [1, 2, 8],
    [7, 3, 0],
    [6, 5, 4],
]
node_inicial = Node(estado, None, None, 0)
imprimir(node_inicial.estado)
nodes_filhos = expandir_node(node_inicial)
for n in nodes_filhos:
  imprimir(n.estado)
```

Vamos gerar soluções iniciais do quebra-cabeça-8 embaralhando a partir do estado final.


```python
import random
def embaralhar(estado_final):
  Z = copiar(estado_final)

  for i in range(0,30):
    operacao = random.randint(1,4) # gera um num. int aleatorio [1,4]
    if operacao == 1:
      C = cima(Z)
      if C != None:
        Z = C

    if operacao == 2:
      C = baixo(Z)
      if C != None:
        Z = C

    if operacao == 3:
      C = esquerda(Z)
      if C != None:
        Z = C

    if operacao == 4:
      C = direita(Z)
      if C != None:
        Z = C

  return Z

estado_final = [
    [1, 2, 3],
    [8, 0, 4],
    [7, 6, 5],
]


estado_inicial = embaralhar(estado_final)
print('Estado inicial')
imprimir(estado_inicial)
print('Estado final')
imprimir(estado_final)

```

Busca em profundidade, com parâmetro de limite da profundidade da busca.

```python
from pythonds.basic.stack import Stack

def busca_profundidade(estado_inicial, estado_final, profundidade_maxima=300):
  node_inicial = Node(estado_inicial,None,None,0)

  pilha = Stack()
  pilha.push(node_inicial)

  node_atual = pilha.pop()
  while(node_atual.estado!=estado_final):
    #imprimir(node_atual.estado)
    #print(node_atual.profundidade)
    node_filhos = expandir_node(node_atual)

    for node_filho in node_filhos:
      if node_filho.profundidade <= profundidade_maxima:
        pilha.push(node_filho)

    if pilha.isEmpty(): # terminou a busca e nao encontrou solucao
      node_atual=node_inicial
      break
    node_atual = pilha.pop()

  caminho_solucao = []
  while(node_atual.pai!=None):
    caminho_solucao.insert(0,node_atual)
    node_atual=node_atual.pai

  return caminho_solucao
```

```python
solucao = busca_profundidade(estado_inicial, estado_final, profundidade_maxima=300)
if len(solucao)==0:
  print('Nao encontrei solucao =(')
```

```python
imprimir(estado_inicial)
for s in solucao:
  print(s.movimento)
  imprimir(s.estado)
```

Busca em Largura

```python
def busca_largura(estado_inicial, estado_final):

    node_inicial=Node(estado_inicial,None,None,0)
    fila=[]
    fila.append(node_inicial)
    node_atual=fila.pop(0)

    while(node_atual.estado!=estado_final):

        node_filhos=expandir_node(node_atual)

        for node_filho in node_filhos:
          fila.append(node_filho)

        node_atual=fila.pop(0)


    caminho_solucao=[]
    while(node_atual.pai!=None):
        caminho_solucao.insert(0,node_atual)
        node_atual=node_atual.pai
    return caminho_solucao
```

```python
solucao = busca_largura(estado_inicial, estado_final)
```

```python
imprimir(estado_inicial)
for s in solucao:
  print(s.movimento)
  imprimir(s.estado)
```
